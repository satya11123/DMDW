```python
import pandas as pd
dataset1=pd.read_csv('Toyota.csv')
```

```python
dataset1
```

```python
dataset1.dropna(axis = 0, inplace=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.hist(dataset1['KM'], color = 'blue', edgecolor = 'white', bins = 5)
plt.title('Histogram of Kilometer')
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(dataset1['Age'],dataset1['Price'],c='blue')
plt.title('Scatter plot of Price vs Age of the Cars')
plt.xlabel('Age (months)')
plt.ylabel('Price (Euros)')
plt.show()
```

```python
counts = [979,120,12]
fuelType = ("Petrol","Diesel","CNG")
index = np.arange(len(fuelType))
plt.bar(index, counts, color=['red', 'blue', 'cyan'])
plt.title("Bar plot of fuel types")
plt.xlabel("Fuel Types")
plt.ylabel("Frequency")
plt.xticks(index,fuelType,rotation = 90)
plt.show()
```

```python
plt.hist(dataset1['Price'], color = 'blue', edgecolor = 'white', bins = 5)
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
